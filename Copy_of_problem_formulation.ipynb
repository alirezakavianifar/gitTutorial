## Decision variables



#### Definitions:

1. **Truck Movement Between Nodes:**
   
   $$ X_{ij}^{tr} =
   \begin{cases}
   1 & \text{if truck } tr \in TR \text{ moves between nodes } i \in N_0 \text{ and } j \in N^- \text{, where } i \neq j \\
   0 & \text{otherwise}
   \end{cases} $$
  

2. **UAV Movement from Truck to Node and Back:**
   
   $$ Y_{ijB}^{tr} =
   \begin{cases}
   1 & \text{if UAV of truck } tr \in TR \text{ departs from node } i \in N_0 \text{, flies to } j \in C \text{, and returns to the end depot or truck at node } B \in N^+ \text{, } \{0, j, B\} \in TP \\
   0 & \text{otherwise}
   \end{cases} $$
   

3. **UAV Specific Route:**
   
   $$ RP_{ijB}^k =
   \begin{cases}
   1 & \text{if UAV } k \in K \text{ of truck departs from node } i \in N_0 \text{, flies to } j \in C \text{, and returns to the end depot or truck at node } B \in N^+ \text{, } \{0, j, B\} \in TP \\
   0 & \text{otherwise}
   \end{cases} $$
   

4. **RPA Delivery Sequence:**
   
   $$ z_{ij} =
   \begin{cases}
   1 & \text{if delivery location } i \text{ is served immediately before } j \text{ by a RPA} \\
   0 & \text{otherwise}
   \end{cases} $$
  

5. **RPA Battery Replacement:**
   
   $$ y_i =
   \begin{cases}
   1 & \text{if RPA battery is replaced after returning from delivery location } i \\
   0 & \text{otherwise}
   \end{cases} $$
   

6. **RPA Assignment to Delivery Location:**
   
   $$ x_{ki} =
   \begin{cases}
   1 & \text{if RPA type } k \text{ is assigned to deliver a package to location } i \\
   0 & \text{otherwise}
   \end{cases} $$
   

7. **Route from Depot to Node:**
   
   $$ x_{Dji} =
   \begin{cases}
   1 & \text{Binary variable indicating whether a route exists from the depot to } i \\
   0 & \text{otherwise}
   \end{cases} $$
   

8. **Route from Node to Depot:**
   
   $$ x_{iDj} =
   \begin{cases}
   1 & \text{Binary variable indicating whether a route exists from } j \text{ to the depot} \\
   0 & \text{otherwise}
   \end{cases} $$
   

9. **RPA Travel to Charging Point:**
   
   $$ Y_{ij} =
   \begin{cases}
   1 & \text{Binary variable indicating whether a RPA travels from point } i \text{ to a charging point } j \\
   0 & \text{otherwise}
   \end{cases} $$
   

#### Example Constraints:

1. **Each truck can move to at most one node at a time:**
   
   $$ \sum_{j \in N^-} X_{ij}^{tr} \leq 1, \quad \forall i \in N_0, tr \in TR $$
  

2. **Each UAV must return to the truck or end depot:**
   
   $$ Y_{ijB}^{tr} = 0, \quad \forall i \in N_0, j \in C, B \in N^+, \{0, j, B\} \notin TP $$
   

3. **UAV routes are consistent:**
  
   $$ \sum_{j \in C} RP_{ijB}^k = \sum_{B \in N^+} Y_{ijB}^{tr}, \quad \forall k \in K, i \in N_0, tr \in TR $$
   

4. **Battery replacement after certain deliveries:**
   
   $$ y_i = 1 \implies \text{Battery replacement after delivery at } i, \quad \forall i \in N $$
   

5. **RPA assignment constraint:**
   
   $$ \sum_{k \in K} x_{ki} = 1, \quad \forall i \in C $$
   

6. **Depot route constraints:**
   
   $$ \sum_{i \in C} x_{Dji} \leq 1, \quad \forall j \in N_0 $$
   
   
   $$ \sum_{j \in N_0} x_{iDj} \leq 1, \quad \forall i \in C $$
   

7. **Charging point travel constraint:**
   
   $$ \sum_{j \in C} Y_{ij} \leq 1, \quad \forall i \in N $$

## 1.1.4. Mathematical Model

Based on the previous definitions, the model and variables can be explained as follows: These parameters include point-specific values, service times, RPA and Truck capacities, operational time windows, cost factors, battery-related parameters, and penalties associated with various operational conditions.



\$ z_1 = \min \left( \sum_{tr \in TR} \sum_{B \in N_+} \sum_{j \in C} \sum_{i \in N_0} y_{ijB}^{tr} \cdot FC_k \cdot dy_{ijB} \cdot C + \sum_{i \in N_0} \sum_{j \in N_+} \sum_{tr \in TR} x_{ij}^{tr} \cdot c_{ij} + \sum_{i \in N_0} \sum_{j \in C} \sum_{B \in N_+} \sum_{k \in K} RP_{ijB}^K fi y_k + W_2 \cdot \sum_{i \in D} E_i + W_3 \cdot \sum_{i \in D} L_i + \sum_{d \in D} z_d \cdot fix_d + \sum_{i \in N_0} \sum_{j \in C} \sum_{B \in N_+} Y_{ijB}^{tr} \cdot C_{ij}^B + \sum_{f \in F} Y_{df} \cdot de_d \cdot cp_f \right) \$

The first objective function, denoted as \(Z_1\), aims to minimize the overall operational cost. This includes costs associated with RPA and Truck transportation, battery usage along routes, fixed RPA usage costs, costs for constructing distribution points at potential locations, penalties for not adhering to soft time windows, and preparatory expenses.

$$

The second objective function, denoted as \(Z_2\), seeks to minimize the service time. Since RPA velocity has different modes according to different routing scenarios while they are distinctive by the level of air traffic conditions. In a way, there is a set \(U\) of scenarios adjusted with possible routes which end in the set of \(j\) nodes. The number of scenarios is determined by experts who investigate the traffic conditions based on factors such as air parcels, severe climate, The National Aviation System (NAS) regulations, and so on, in the possible routes from nodes \(i\) to \(j\).

$$
Z_3 = \min \max \left( s_{ij}^{tr} + \sum_{j \in N_0, tr \in TR} (Y_{ijB}^{tr} + x_{ij}) + s_{ij} \sum_{i \in N_0, U \in K} RP_{ijB}^k \right)
$$


The third objective function, denoted as \(Z_3\), is formulated to minimize the risk index. It is calculated as the maximum of the sum of routing risk imposed on each pair of nodes \(i\) and \(j\), where \(i\) and \(j\) belong to the union of \(P\) and \(D\) (i.e., PUD). The risk index is determined by the sum of the weighted components (factors) of routing risk for each node pair. The applied weighting method has been inspired from the (Mahmoodi et al. 2022) work. In their study, SORA Standard has been applied to estimate the imposed risk index. It was created and developed by JARUS (Joint Authorities for Rulemaking of Unmanned Systems). Concerning SORA methodology, there are multiple risks to consider in a RPAS operation, but they can be classified into ground risks and in terms of safety. Ground risks are specifically those involving third parties in the ground, whereas air risks are those involving third parties in the air.


$$
Z_4 = \min \left( \sum_{i \in C_E} \left( c_i \cdot \sum_{j \in N_0, tr \in TR} C_{ij}^{tr} x_{ij}^{tr} + \left( c_k + C_l \cdot c_t \right) \sum_{k \in N_0, j \in N} Y_{ij}^{tr} + Y_{ij} \cdot d_{ij} \cdot C_{bat} \left( \sum_{i} Y_i \right) \right) \right)

Constraints in the model serve various functions to optimize the RPAS logistics system. Constraints (5) to (8) deal with RPAS allocation and route efficiency: Constraint (5) allows flexible allocation of RPAS and Truck to delivery service nodes, (6) mandates serving specific points, (7) enables RPAS to serve multiple distributors, and (8) prevents self-connecting routes. (8a) dictate that all trucks must depart from and return to the same depot at most once. Constraint (8b) serves to prevent sub-tours for each truck, and constraint (8c) imposes the condition that a truck that visits node j must also leave from the same node. Constraints (9) to (11) and (13) to (14) manage time and capacity: (9) and (10) set service start times, (11) fixes RPAS and Truck start time at zero, (13) computes initial load, and (14) assesses loading capacity. (14a) requires a truck to accompany a RPA to both its launch and rendezvous locations.
